```python
df['Family_size']=df['Parch']+ df['SibSp'] + 1
df
```

```python
df.drop(columns=["Cabin","SibSp", "Parch"], inplace=True)
df
```

```python
df.drop(columns=["PassengerId"], inplace=True)
df
```

```python
# Extract titles from passenger names to create some sort of new column which we think could influence survival
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df
```

```python
df.drop(columns=["Name"], inplace=True)
```

```python
df
```

```python
# Create age groups
bins = [0, 18, 30, 50, float('inf')]
labels = ['Child', 'Young Adult', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
df
```

```python
# Group data by age group and calculate survival statistics
age_group_survival = df.groupby('AgeGroup')['Survived'].mean()

# Create a bar chart
plt.bar(age_group_survival.index, age_group_survival.values)
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Age Group')
plt.show()
```

```python
# Calculate survival rate by passenger class
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(survival_rate_by_class.index, survival_rate_by_class.values * 100, color='red')

# Add labels and title
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate (%)')
plt.title('Survival Rate by Passenger Class')

# Customize x-axis ticks and labels
plt.xticks(survival_rate_by_class.index, labels=['1st Class', '2nd Class', '3rd Class'])

# Display the plot
plt.show()
```

```python
import seaborn as sns
corr_df = df[['Age', 'Family_size','Survived','Fare']].copy()

corr_matrix = corr_df.corr()

#using a heatmap to visualize
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
```

```python
plt.figure(figsize=(10, 6))
plt.violinplot(df['Fare'], vert=False, widths=0.7, showmeans=True, showextrema=True, showmedians=True)

# Add labels and title
plt.xlabel('Fare')
plt.ylabel('Distribution')
plt.title('Distribution of Fares Paid by Passengers')
```

```python
# Calculate survival rates per title and pclass
title_survival_rates = (
    df.groupby(["Sex", "Title"])["Survived"]
    .mean()
    .unstack()
    .fillna(0)  # Fill missing values with 0 for consistent display
)

# Create the bar chart
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
title_survival_rates.plot(kind="bar")

# Customize chart elements
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.title("Survival Rates by Passenger Title and Sex")
plt.legend()
plt.xticks(rotation=0)  # Adjust x-axis label rotation as needed
plt.grid(True)

# Visual cues for survival rates
plt.axhline(y=0.5, color="gray", linestyle="--")  # Add a dashed line at 50% survival rate
plt.ylim(0, 1)  # Set y-axis limits to 0-1 for survival rates

plt.show()
```
